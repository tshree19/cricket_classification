```python
import os
import random
import torchaudio
import librosa
```

```python
#create a file with names of cricket audio file: raw data (later to be used to shuffle and use them as train and test data)
wav_dir = "/content/drive/MyDrive/Cricket Song Files/wav Tanu/"
wav_files_list = sorted(os.listdir(wav_dir))
datafile_path = r'/content/drive/MyDrive/Cricket_song_classification_and_analysis/data.txt'
with open(datafile_path, 'w') as data:
  for item in wav_files_list:
    data.write(item + "\n")
```

```python
print(len(wav_files_list))
```

```python
datafile = datafile_path
with open(datafile, 'r') as file:
  lines = [line.strip() for line in file.readlines()]

print(lines)
print(len(lines))
Genus_dict = {}
for line in lines:
  genus = line.split(" ")[2]
  if genus in Genus_dict:
    Genus_dict[genus].append(line)
  else:
    Genus_dict[genus] = [line]

print(Genus_dict)
```

```python
print(len(Genus_dict))
```

```python
genus_count = {}
for genus,item in Genus_dict.items():
  genus_count[genus] = len(item)
print(genus_count)

```

```python
min_genus = 30
genus_list = [(genus, count) for genus,count in genus_count.items() if count >= 30 ]
print(genus_list)
# print(each_count)
```

```python

waveform_list = {}
wav_list = {}
for wav_file in wav_files_list:
  # print(wav_file)
  genus_name = wav_file.split(" ")[2]
  # print(genus_name)
  wav, sample_rate = torchaudio.load(os.path.join(wav_dir, wav_file))
  duration = float(wav.shape[1]) / sample_rate
  if duration >= 15 and wav_file.find("(MCL)") != -1 and genus_count[genus_name] >= 30: #MCL files have approx 5 secs of voiceover
    wav_list[wav_file] = duration
    waveform_list[wav_file] = wav
  elif duration >= 10 and wav_file.find("(SINA)") != -1 and genus_count[genus_name] >= 30:
    wav_list[wav_file] = duration
    waveform_list[wav_file] = wav
```

```python
print(wav_list)
print(waveform_list)
print(len(waveform_list))
```

```python
datafile_path = r'/content/drive/MyDrive/Cricket_song_classification_and_analysis/raw_data/data.txt'
with open(datafile_path, 'w') as data:
  for key,item in wav_list.items():
    data.write(key + "\n")
```
